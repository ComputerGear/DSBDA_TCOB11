```python

!pip install seaborn pandas matplotlib --quiet

# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print("First 5 rows of the Titanic dataset:")
print(titanic.head())

# Plot histogram of fare prices
plt.figure(figsize=(8, 5))
sns.histplot(titanic['fare'], bins=30, kde=True, color='skyblue')

# Labeling the plot
plt.xlabel('Fare Price')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Ticket Prices (Fare) on Titanic')

# Show the plot
plt.show()
```

```python
# Dist-Plot for Fare
plt.figure(figsize=(8, 6))
sns.histplot(titanic['fare'], kde=True, color='blue', bins=30)
plt.title("Distribution of Fare with KDE")
plt.show()
```

```python
# Joint Plot for Fare vs Age
plt.figure(figsize=(8, 6))
sns.jointplot(x='fare', y='age', data=titanic, kind='scatter', color='green')
plt.show()
```

```python
# Rug Plot for Fare
plt.figure(figsize=(8, 6))
sns.rugplot(titanic['fare'], color='purple')
plt.title("Rug Plot of Fare")
plt.show()
```

```python
# A. Bar Plot - Shows the average fare for each passenger class
sns.barplot(x='class', y='fare', data=titanic)
plt.title("Average Fare by Class")
plt.show()
```

```python
# B. Count Plot - Shows the number of passengers in each class
sns.countplot(x='class', data=titanic)
plt.title("Passenger Count by Class")
plt.show()
```

```python
# C. Box Plot - Shows age distribution for each class
sns.boxplot(x='class', y='age', data=titanic)
plt.title("Age Distribution by Class")
plt.show()
```

```python
# D. Violin Plot - Shows the distribution of fares for each class
sns.violinplot(x='class', y='fare', data=titanic)
plt.title("Fare Distribution by Class")
plt.show()
```

```python
# a. Strip Plot - Shows fare distribution for each class
sns.stripplot(x='class', y='fare', data=titanic, jitter=True)
plt.title("Strip Plot: Fare Distribution by Class")
plt.show()
```

```python
plt.figure(figsize=(8, 6))  # Increase figure size for better visualization
sns.swarmplot(x='class', y='fare', data=titanic.sample(250), size=1.5)  # Reduce marker size & use sample
plt.title("Swarm Plot: Fare Distribution by Class (Sampled Data)")
plt.ylim(0, 300)  # Limit y-axis to remove extreme fares
plt.show()
```

```python
# a. Heat Map (Correlation between numeric features)
plt.figure(figsize=(8, 6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Heatmap: Correlation Between Features")
plt.show()
```

```python
# b. Cluster Map (Groups similar data)
sns.clustermap(titanic.corr(numeric_only=True), cmap='coolwarm', annot=True)
plt.show()
```

```python
#From the chapter 10
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')
```

```python
# Scatter Plot: Sepal Length vs Sepal Width
#Scatter Plot: Displays relationship between two continuous features.
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris,hue='species')
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
# Strip Plot: Sepal Length by Species
#Strip Plot: Shows individual data points.
sns.stripplot(x='species', y='sepal_length', data=iris, jitter=True)
plt.title("Strip Plot: Sepal Length by Species")
plt.show()
```

```python
# Swarm Plot: Sepal Length by Species
#Swarm Plot: Avoids point overlap by adjusting positions.
sns.swarmplot(x='species', y='sepal_length', data=iris)
plt.title("Swarm Plot: Sepal Length by Species")
plt.show()
```

```python
# Count Plot: Number of Samples per Species
#Count Plot: Shows counts of categories.
sns.countplot(x='species', data=iris)
plt.title("Count Plot: Number of Samples per Species")
plt.show()
```

```python
# Box Plot: Sepal Length by Species
#Box Plot: Displays distribution based on quartiles.
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Box Plot: Sepal Length by Species")
plt.show()
```

```python
# Violin Plot: Sepal Length by Species
#Violin Plot: Combines box plot and density plot.
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title("Violin Plot: Sepal Length by Species")
plt.show()
```

```python
# Pair Plot: All Features
#Pair Plot: Shows relationships between all features.
sns.pairplot(iris, hue='species')
plt.title("Pair Plot: All Features")
plt.show()
```

```python
# Pie Chart: Proportion of Species
#Pie Chart: Displays proportions of species.
species_count = iris['species'].value_counts()
plt.pie(species_count, labels=species_count.index, autopct='%1.1f%%', colors=sns.color_palette('Set2', len(species_count)))
plt.title("Pie Chart: Proportion of Species")
plt.show()
```
