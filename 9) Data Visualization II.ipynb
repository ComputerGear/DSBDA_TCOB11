```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv(r"C:\Users\LOQ\Downloads\titanic.csv")
```

```python
# Create Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title('Distribution of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
### Observations:
#- Males had a wider and older age range than females.
- Female survivors were mostly younger adults.
- More children and younger females survived.
- Older males had the lowest survival rate.
- The policy "women and children first" is evident from the boxplot.
```
